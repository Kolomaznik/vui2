### Čo je NLP?


Spracovanie prirodzeného jazyka (NLP) je oblasť umelej inteligencie, ktorá umožňuje počítačom analyzovať a porozumieť ľudskému jazyku pri komunikácií s človekom. Aby sme mohli komunikovať s počítačom, počítač potrebuje rozumieť našemu jazyku. Preto vznikol NLP, čo je proces, ktorý naučí počítač rozumieť, analyzovať a manipulovať ľudský jazyk.
 
NLP kombinuje AI s výpočtovou lingvistikou a počítačovou vedou na spracovanie ľudských alebo prírodných jazykov a reči. Proces možno rozdeliť na niekolko častí. Prvá a druhá časť NLP je syntaktická a sémantická analýzu. Cieľom NLP je porozumieť prirodzenému jazyku prijatému počítačom. Počítač používa vstavaný štatistický model na vykonávanie rutiny rozpoznávania reči, ktorá prevádza prirodzený jazyk na programovací jazyk.  Výstup alebo výsledok v textovom formáte štatisticky určuje slová a vety, ktoré boli s najväčšou pravdepodobnosťou povedané. Táto prvá úloha sa nazýva proces reči na text. Po týchto procesoch počítač pravdepodobne teraz chápe význam hovorenej reči.
Dalším krokom NLP je prevod textu na reč. V tejto fáze sa počítačový programovací jazyk prevádza pre používateľa do zvukového alebo textového formátu. Napríklad finančnému chatbotu, ktorý dostane otázku typu „Ako je na tom Google dnes?“ s najväčšou pravdepodobnosťou skontroluje online finančné stránky na akcie spoločnosti Google a môže sa rozhodnúť, že ako odpoveď vyberie iba informácie, ako je cena a objem.

NLP sa pokúša dosiahnuť, aby boli počítače inteligentné tým, že ľudí presvedčí, že interagujú s iným človekom. Turingov test, ktorý navrhol Alan Turing v roku 1950, uvádza, že počítač môže byť plne inteligentný, ak dokáže premýšľať a konverzovať ako človek bez toho, aby vedel, že hovorí so strojom. Testom zatiaľ prešiel iba jeden počítač – chatbot s osobnosťou 13-ročného chlapca. To neznamená, že inteligentný stroj nie je možné zostrojiť, ale načrtáva ťažkosti, ktoré sú spojené s nutením počítača premýšľať alebo konverzovať ako človek. Pretože slová sa dajú použiť v rôznych kontextoch a stroje nemajú skúsenosti z reálneho života, ktoré majú ľudia na vyjadrovanie a opisovanie entít slovami, môže chvíľu trvať, kým svet úplne odstráni počítačový programovací jazyk.

## Komponenty NLP
* Lexikálna analýza rozdeľuje text na paragrafy, vety a slová. Táto analýza zahŕňa aj identifikáciu a analýzu štruktúry slov.

* Pragmatická analýza sa zaoberá komunikáciou a interpretáciou jazyka.Zaoberá sa odvodením zmysluplného jazyka v rôznych situáciách.

# Spracovanie a očistenie textu
Po získaní dát je ako druhý krok spracovanie a očistenie dát. Na to aby sme z dát získali čo najviac inforacií je potrebné text očistiť od slov, ktoré nenesú veľa informácií. 

Medzi očistenie dát patrí mnoho rôznych spôsobov. Medzi najčastejšie patrí odstránenie znamienok, zmena všetkých písmen na malé, odstránenie čísel, stemming, lematizácia, odstránenie stop words, odstránenie nezmyselného textu a mnoho dalších.  

## 1.   Otvorenie súboru pre prácu


```python
#Otvorenie dokumentu :
text_file = open("Natural_Language_Processing_Text.txt")

#Prečítanie súboru :
text = text_file.read()

#Vypísanie textu :
print(text)
print("\n")
#Dĺžka textu :
print (len(text))
```


## 2.   Načítanie knižníc a tokenizácia
Tokenizácia rozdelí text na menšie kúsky. Text môžeme rozdeliť na vety, slová, bi-gramy (dvojice slov).




```python
#Importovanie potrebných knižníc:
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
#Tokenizácia textu na vety :
sentences = sent_tokenize(text)

#Počet viet :
print (len(sentences))

#Vety :
sentences
```

```python
#Tokenizácia textu na slová :
words = word_tokenize(text)

#Počet slov :
print (len(words))

#Slova :
words
```

## 3.   Zmena veľkých písmen na malé a odstránenie znamienok
Aby sa správne určovala frekvencia slov je potrebné aby veľkosť písmeniek bola rovnaká a aby sa odstránili znamienka aby medzi najčastejšie slová nepatrila napríklad ".".


```python
#Prázdny list na uloženie slov:
words_no_punc = []

#Odstránenie interpunkčných znamienok a zmena na písmen na malé:
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Výpis slov bez interpunkčných znamienok:
print (words_no_punc)

print ("\n")

#Počet slov :
print (len(words_no_punc))
```



## 4. Frekvencia slov a graf frekvencie slov




```python
#Import potrebných knižníc:
from nltk.probability import FreqDist

#Distribúcia frekvencie :
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

```python
#Plot the graph for fdist :
import matplotlib.pyplot as plt

fdist.plot(10)
```

## 5. Odstránenie stop words
Stop words sú slová, ktoré sa v texte nachádzajú najčastejšie a vetám nedodávajú skoro žiadnu váhu a význam. Sú to slová, ktoré fungujú ako mosty a ich úlohou je zabezpečiť aby boli vety gramaticky správne. Odstraňovanie stop words je jedným z najbezžnejšich krokov predspracovania textu. V NLP nájdeme aj úlohy ako klasifikácia a získavanie informácií, ktoré nie sú ovplyvňované týmito slovami a preto nie je potrebné ich odstraňovať. Tieto slová sú v zozname, ktorý bol ručne vytvorený a líši sa pre každý jazyk. Knižnica NLTK má vlastný zoznam stop words.

```python
#Importovanie stopwords
from nltk.corpus import stopwords

#List stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

```python
#Prázdny list pre uloženie očistených slov:
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
        
print(clean_words)
print("\n")
print(len(clean_words))
```

```python
#Frekvencia slov po odstránení stop words
fdist = FreqDist(clean_words)

fdist.most_common(10)

fdist.plot(10)
```

## 6. Stemming
Slová sa väčšine jazykov majú viacero tvarov a nám na stačí len koreň slova. Stemming je proces odstraňovania koncoviek slova aby sme dostali koreň slova. Existuje niekoľko druhov stemmerov, my budeme v príklade používať Porter, ktorý je najpoužívanejší. Za zmienku určite stojí aj Snowball stemmer, ktorý umožňuje pracovať s rôznymi jazykmi ale pre nagličtinu dá rovnaký výsledok ako Porterov stemmer.

```python
#Import stemmingov knižnice :
from nltk.stem import PorterStemmer

porter = PorterStemmer()

#Word-list pre stemming:
word_list = ["plays","decreases","Studies","Studied", "leaves", "singing","caring","was"]

for w in word_list:
    print(porter.stem(w))
```

## Chunking
Chunking sa používa na zhromaždenie jednotlivých informácií a ich zoskupenie do väčších častí viet. Extrahuje zmysluplné frázy z neštrukturovaného textu. Tento krok prebieha po PoS. Chunking znamená doslova skupina slov, ktorá rozdeľuje jednoduchý text na slovné spojenia, ktoré sú zmysluplnejšie ako jednotlivé slová. V príklade si ukážeme extrahovnaie Noun Phrade (podstatná fráza alebo nominálna fráza), ktorá má fuknciu ako podstatné meno. Je to fráza, ktorá má ako "hlavu" podstatné meno spojené s determinantami alebo prídavnými menami.

```python
#Extrahovanie Noun Phrase z text :

# ? - voliteľný znak
# * - 0 alebo viac opakovaní
grammar = "NP : {<DT>?<JJ>*<NN>} "

#Creating a parser :
parser = nltk.RegexpParser(grammar)

#Parsing text :
output = parser.parse(tagged_words)
print (output)
```

# Bag of words
Je zjednodušená reprezentácia textu pri NLP. Je to metóda extrahovania  základných slov z textu aby sme ich mohli použiť pre modely strojového učenia. V procese zahodíme poradie slov. Je to teda súhrn slov, ktorý predstavuje vetu s počtom slov, kde poradie nie je relevanté. Používa sa pri klasifikácií alebo ked je potrebné aby stroj moholo čítať slová ako čísla.

BoW.png

Z pôvodného textu odstránime nepotrebné veci ako interpunkčné znamienka, stop words. Tento očistený text tokenizujeme a rozdelíme na slová. Na základe toho vytvoríme slovník, ktorý obsahuje slová z textu bez stop words. Následne vygenerujeme slovník, ktorý bude obsahovať slová a ich frekvenciu výskytu vo vetách.








```python
#Import potrebných knižníc :
from sklearn.feature_extraction.text import CountVectorizer

#Text pre analýzu :
sentences = ["Jim and Pam travelled by the bus:",
             "The train was late",
             "The flight was full.Travelling by flight is expensive"]

#Vytvorenie objektu :
cv = CountVectorizer()

#Generovanie výstupu pre Bag of Words :
B_O_W = cv.fit_transform(sentences).toarray()

#Všetky slová s ich indexom v modeli:
print(cv.vocabulary_)
print("\n")

#Features:
print(cv.get_feature_names())
print("\n")

#Výpis výstupu :
print(B_O_W)
```

# TF-IDF
TF-IDF (Term frequency - Inverse Dense Frequency)je technika ktorá využíva štatistické výpočty. Používa sa, ked chceme zistiť význam vety. Táto technika funguje tak, že hodnotí ako dôležité je slovo pre dokument v zbierke dokumentov. Pri používaní Bag of worlds je vo vete "Today is a beuatiful day" slovo "a" rovnako dôležité ako "beautiful" čo ale vo význame nie je pravda. Vdaka TF-IDF ale vieme vyrátať, že slovo "a" sa nachádza veľakrát vo viacerých dokumentoch a preto má nízku hodnotu. 

Pre výpočet potrebujeme vyrátať TF frekvenciu slova a IDF Inverse document frequency. 
Frekvenciu slova TF vyrátame ako Frekvencia slova vo vete/ Velkový počet slov vo vete.
IDF znamená ako bežné ale jedinečné je slovo v skupine dokumentov. ćím je výsledok bližsie k 0 tým je slovo bežnejšie. Túto metriku môžeme vypočítať ako celkový počet dokumentov/ počet dokumentov v ktorých sa dané slovo nachádza a nájsť logaritmus daného čísla. 
Vynásobením týchto dvoch čísel získame TF.IDF. ćím vyššie je skóre čísla, tým významnejšie je dané slovo v dokumente.

idfpng.png

## Čištění dat

Pokud pracujeme s číselnými daty, čištění dat většinou obsahuje odstraňování null hodnot, duplicitních dat, zpracování extrémů, atd. U textových dat používáme odlišné metody, které se nazývají jako pre-procesní techniky (pre-processing techniques).

Textová data se dají stále více a více očišťovat, technik existuje několik a data je možné stále více a více upravovat.
Jedním z doporučovaných přístupů u čištění dat je tzv. MVP přístup (minimum viable product) - začít jednoduchý kroky a postupně data očisťovat v iteracích, dokud nemají strukturu, kterou požadujeme.

Tady je seznam obvyklých technik čištění dat.

**Techniky pro čistění celého textu:**

- Přeměnit celý text na malá písmena
- Odstranění interpunkce
- Odstranění číselných hodnot
- Odstranění nesrozumitelného textu (např. /n)
- Tokenizace textu
- Odstranění tzv. stop words

**Další možné kroky po provedení tokenizace:**

- Stematizace (nalezení kořene) / lematizace (různé tvary slov jsou brány jako jedno (např. jízda, jízdy, jízd,..))
- Určení slovních druhů
- Vytvoření tzv. bi-grams or tri-grams
- Odstranění překlepů
- Atd..

Vysvětlení tokenizace:
    - rozdělit text na menší částí, tzv. tokeny. Nejčastěji je tokenem slovo (ale také věta,...)
    
Vysvětlení stop words:
    - slova, které obsahují velmi málo významu (a, the, on, out)
    
Vysvětlení bag of words:
    - kolekce slov, kde nezáleží na jejich pořadí

```python
# čištění dat
import re
import string

def clean_text_round1(text):
    '''lowercase, odstranit text v hranatých závorkách, odstranění interpunkce and odstranění slov obsahující čísla.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# očištěná data
data_clean = pd.DataFrame(data_df.transcript.apply(round1))
data_clean
```

```python
# druhé kolo čištění dat
def clean_text_round2(text):
    '''Odstranění dalších nesrozumitelných znaků'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

## Změna sentimentu v průběhu času

Podíváme se, jak se měnil sentiment jednotlivých vystoupení v průběhu času.

```python
# Každý vystoupení rozdělíme na 10 částí

import math

def split_text(text, n=10):
    # výpočet délky textu, velikost každé části textu a kde každá část začíná
    length = len(text)
    size = math.floor(length / n)
    start = np.arange(0, length, size)

    split_list = []
    for piece in range(n):
        split_list.append(text[start[piece]:start[piece]+size])
    return split_list
```

```python
# list všech rozdělených částí všech vystoupeních
list_pieces = []
for t in data.transcript:
    split = split_text(t)
    list_pieces.append(split)
    
list_pieces
```

### Latent Dirichent Allocation

latent: schovaný

dirichent: typ rozdělení pravděpodobnosti

allocation: rozdělení

**myšlenka LDA:** každý dokument se skládá z několika témat, každé téma je definováno několika slovy


Řekněme, že máme 500 různých slov a 100 různých dokumentů. Každý dokument obsahuje průměrně 200 slov. Jak určíme, do jaké kategorie každý dokument patří? Jedním ze způsobů je propojit každý dokument se slovy která obsahuje, jak je naznačeno na obrázku:



Z toho pak můžeme usoudit, že pokud dva dokumenty jsou napojené na podobná slova, mluví o podobném tématu. Ale není to úplně dobré řešení. Museli bychom si ten dokument přečíst, abychom věděli jaké má téma a z toho pak usoudili, že dokumenty obsahující podobná slova mluví o podobném tématu.

Jak by se toto dalo lépe vyřešit je pomocí tzv. latentní (skryté) vrstvy. Řekněme, že víme, že v dokumentech se opakuje 5 různých témat. Ale my nevíme, jaké téma je v jakém dokumentu, pozorujeme pouze slova v dokumentu. Proto se této mezivrstvě říka skrytá. Tudíž co uděláme je to, že spojíme slova do těch témat podle toho, jak každý slovo se do toho tématu hodí a pak podle toho, jaká slova každý dokument obsahuje, tak jim přiřadíme témata.







K tomu, abychom mohli techniku Topic modeling provést, potřebujeme DTM (document-term matrix) a počet témat, které chceme, aby algoritmus rozlišil.

Jakmile tuto techniku začneme nad daty provádět, je potřeba jednotlivé výsledky interpretovat a určit, jestli mix různých slov v daném tématu dává smysl. Pokud ne, můžeme změnit počet témat, které hledáme, počet termů v dokumentu, změnit parametry, nebo případně vyzkoušet jiný model.

Jeden dokument může mít více témat zahrnutých v sobě a jak jsou tyto témata rozdělena je právě určeno rozdělením pravděpodobností (dirichent).

**Příklad:** 

Můžeme hledat 4 různé témata a jedno téma bude určené slovy: (0.3\*Auto, 0.4\*Žvýkačka, 0.2\*Láska, 0.1\*Guláš)
Z toho nejspíš usoudíme, že toto téma není dobře určené a zkusíme tedy jiné parametry.

Pak nám může vyjít: (0.3\*Kočka, 0.4\*Pes, 0.2\*Věrný, 0.1\*Zlý). Z toho usoudíme, že tohle téma by mohlo mít název *Zvířata*.

### Jak LDA funguje

Cíl: Chceme, aby LDA určil pro každý dokument témata a pro každé téma slova, která jej definují.
1. zvolíme počet témat, které chceme hledat v dokumentech
2. náhodně každému slovu přiřadí nějaké téma
3. LDA prochází každé slovo a updatuje přiřazení: příklad - slovu guláš je přiřazeno téma A(zvířata) -> podívá se, jak moc se to přiřazené téma A vyskutuje v dokumentu. A pak se podívá, jak moc často se to slovo vyskytuje v tématu A. Podle těchto hodnot pak slovu guláš případně přiřadí jiné téma.
4. třetí krok se opakuje několikrát

# 3. Generování textu

**input:** corpus (na pořadí záleží)

**output:** nové vtipné věty

Pro generování textu využijeme tzv. **Markov chains**.

**Markov chains** svým způsobem reprezentují, jak se systém mění v průběhu času.
Myšlenkou je, že další proces zálěží pouze na procesu přechodzím.



Pokud se nacházíme ve stavu že svítí slunce, pak je 10% šance, že začne pršet a 90% šance, že počasí zůstane stejné.

Pokud se nacházíme ve stavu že prší, pak je 50% šance, že začne svítit slunce a 50% šance, že bude stále pršet.

**Markov chains** pro text:

Každé slovo je bráno jako stav , ze kterého může přejít do dalších stavů podle určitých pravděpodobností. Jinými slovy, s jakou pravděpodobností bude za daným slovem následovat jiné slovo.





Reprezentace v pythonu pomocí dictionary:

{'hi': ['everyone', 'there', '.'], 'everyone': ['is', 'says'], 'says': ['hi'], ...}

Tohle je jednoduchý způsob generování textu, mnohem složitější způsob je tzv. Long Short-Term Memory (LSTM), kde následující slovo není závislé jen na tom předhcozím, ale na několika předchozích slovech.

## Funkce Markov chains

Klíče by měly být všechny slova v corpusu.
Hodnoty by měly být seznam slov, které se nacházejí za tím slovem, co je klíč.

```python
from collections import defaultdict

def markov_chain(text):
    # tokenizace textu včetně interpunkce
    words = text.split(' ')
    
    # inicializace
    m_dict = defaultdict(list)
    
    # naplnění dat
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)

    # převod na formát dictionary
    m_dict = dict(m_dict)
    return m_dict
```

```python
# vytvoření slovníku konkretního komika
comedian_dict = markov_chain(text)
comedian_dict
```

## Vytvoření generátoru textu

Vytvoříme funkce generující věty. Jako input bude mít:

- slovník, který jsme právě vytvořili
- počet slov, které chceme generovat

```python
import random

def generate_sentence(chain, count=15):

    # chceme, aby první slovo ve větě bylo velké
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()

    # generování následujícího slova na základě předchozího
    for i in range(count-1):
        word2 = random.choice(chain[word1])
        word1 = word2
        sentence += ' ' + word2

    # zakončení věty tečkou
    sentence += '.'
    return(sentence)
```
