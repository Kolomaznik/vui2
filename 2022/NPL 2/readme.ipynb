# **NLP**

NLP neboli "natural language processing" v češtině poté zpracování přirozeného jazyka je soubor technik, které se věnují analýze, generování textů a mluveného slova, které vyžadují určitou (ne nutně absolutní) míru porozumnění přirozenému jazyku strojem. Mezi nejznámější aplikace NLP patří:

První pokusy s NLP má na svědomí Alan Turing, který v roce 1950 publikoval článek "Computing Machinery and Intelligence", který navrhoval Turingův test inteligence, jež obsahoval úkol, kdy si musel poradit s interpretací a genrování přirozené řeči.

**1990:** začátek statistických metod v NLP, velký rozvoj automatických překladačů (IBM Research). Překladače si zatím uměly poradit pouze s omezenou oblastí, na kterou byly nachystány.

**2000:** rozvoj unspurvised algoritmů (algrotimy, které se mohou samy učit z dat, které nebyly ručně anotovány).

# Vektorizace za pomoci tradičních metod


Již od počátku dob zpracování přirozeného jazyka za pomoci počítačů, si pokládáme otázku jak **nejefektivněji reprezentovat data** tak, aby se s nimi (pro stroje) dalo vhodně pracovat, protože algoritmy dokáží pracovat pouze s číselnými hodnotami. Velkým průlomem se stala **vektorizace**, nebo-li reprezentace textu jako vektoru s reálnými hodnotami. Tento přístup převádí slova a dokumenty do číselných vektorů, které jsou výhodné pro zpracování v ML na strojích. 

V momentální době existuje spousta způsobů, jak provádět vektorizaci. Můžeme začít na technikách využívajících naivních výskytu prvků (binárních termínů) až po ty pokročilé reprezentace prvků, které jsou schopny pracovat i s kontextem slov. 

Jedním ze základních nástrojů je **CountVectorizer**. Jedná se o skvělý nástroj z knihovny sci-kit-learn v Pythonu. Tato technika dokáže transformovat zadaný text do vektoru na základě toho, jak často se vyskytuje v celém textu. CountVectorizer vytvoří **matici**, ve které je každé jedinečné slovo reprezentováno sloupcem matice a každý textový vzorek z dokumentu je řadek v matici. Matice je vyplněna hodnotami reprezentujícími, jak často je slovo uvedeno v konkrétním textu.

**TF-IDF** je další z nástrojů, které pro Python poskytuje knihovna scikit-learn. Jedná se o velmi běžný algoritmus pro transformaci textu na smysluplnou reprezentace čísel, kterou jsme schopni přizpůsobit i pro strojové algoritmy založených na **predikci**. Jak již název napovídá, jedná se o techniku složenou ze 2 termínů. **TF** = jedná se o počet výskytů slova ve věte. **IDF** = Je definovaný jako počet celkových dokumentů dělený dokumenty, ve kterých se slovo vyskytuje. Tato technika využívá 3 kroků. Nejprve si vytvoří slovní zásobu jedinečných slov (stejně jako CountVectorizer). Následně vytvoří pole nul pro každou větu v korpusu s velikostí rovnou počtu jedinečných slov v korpusu. Následně se vypočítá TF-IDF pro každé slovo ve větě. Stejně se postupuje pro každé slovo ve slovní zásobě (ve slovníku). Hodnoty ještě můžeme dále normalizovat. Nejčastěji se setkáme s parametrem algoritmu využívající normalizaci l2.  

**Word2Vec** je kombinací modelů používaných k reprezentaci slov v korpusu. Jendá se o sadu modelů neuronových sítí, které chtějí reprezentovat slova ve vektorovém prostoru. Tyto modely dokáží mít velmi vysokou účinnost a fungují dobře i při **porozumění kontextu a vztahů mezi slovy**. Slova co jsou podobná nebo mají nějaký vztah umísťují blízko vedle sebe ve vektorovém prostoru a ty nepodobná jsou umístěna daleko od sebe. V této tříde je využíváno 2 základních modelů. **CBOW** reprezentuje neuronovou síť, která se podívá na okolní slova a předpovídá slovo, které je mezi nimi. **Skip-gram** je neuronová síť, která přijme slovo a následně se snaží předpovědět okolní slova.

Nejprve si vysvětlíme **základní pojmy** využité v této oblasti.

- **Předzpracování textu:** Jelikož data mohou být z různých zdrojů (články, uživatelské komentáře, ...) a mohou být napsány v přirozeném jazyce, který je dobře srozumitelný pro lidi, ale počítač ho zpracovává velmi těžko. 

- **Korpus:** Jedná se o sbírku textů obvykle zaměřenou na určíté téma. V kontextu machine learningu se dá korpus chápat jako vstupní dataset.

- **Normalizace textu:** Vstupní data při zpracování přirozeného jazyka mohou být napsána různými způsoby ale vyjadřovat to stejné. Někteří uživatelé píšou s diakritikou, jen malá písmena atd... Proto je potřeba tyto slova sjednotit. 

- **Stop words:** Jsou slova, které jsou nedílnou součástí přirozeného jazyka, běžně se využívají, ale nemají žádnou přidanou hodnotu pro význam textu. Nejčastěji se tam zahrnují spojky nebo členy. 

- **Tokenizace:** Tento proces slouží pro počítač na rozdělení textů na uchopitelné časti, se kterými je schopný pracovat. Soubor se rozdělí na soubor tokenů. Nejčastějí se využívá token na jedno slovo.

- **N-gramy:** Speciálním připadem pro tokenizaci. Jedná se o n-tici tokenů, v případě textů se jedná o dvě slova po sobě (pro bigramy). Výhodou je zachování vztahů mezi slovy. 

- **Stemming:** Operace se provádí nad každým jedním slovem (proto se nejčastěji provádí až po tokenizaci). Výsledkem stemmingu je nalezení kořene slova bez jakýchkoliv jazykových vlivů (odstranění předpon, přípon, ...). Dalším krokem by mohla být **lemming**. Ta nahrazuje slovo do jeho podoby v prvním pádu.

# Příklad vektorizace za pomoci tradičních metod

Algoritmy strojového učení nemohou pracovat s neupravenými textovými daty. Obecně se využívají ve strojovém učení číselná data, na kterých můžeme provádět jakékoliv matematické operace. Nejčastějším způsobem číselné reprezentace pro tyto typy úloh jsou ve formě vektoru. Množství technik převodu textových dat do numerické reprezentace je spoustu. Nejčastější z nich jsou:

- **Bag of Words**
- Term Frequency and Inverse Document (**TF-IDF**)
- **Word2Vec**

**Předzpracování textu**

Před tím než začneme s daty pracovat, je potřeba je upravit do vhodné podoby. Nezpracovaná data mohou obsahovat číselné hodnoty, interpunkce, speciální znaky, atd. Takovéto hodnoty mohou brzdit výkon modelu, zvyšovat nároky na paměť a přitom nepřinesou žádné zlepšené výsledky. Na nezpracovaná data použijeme řadu technik, které je převedou na smysluplná data a tyto techniky se souhrně nazývají předzpracování textu. Provádějí se následující způsoby:

- **Odstranění hlučných dat:** Za hlučná data můžeme definovat různá záhlaví, zápatí textových souborů, HTML, XML a jiné značkovací znaky. Tyto typy dat nepřinášejí žádnou hodnotu/informaci pro význam zprávy a nedává smysl s nimi pracovat a proto se jich zbavujeme. Pro jejich odstranění je možné v Pythonu využívat různé knihovny a nebo využít regulárních výrazů. 

- **Tokenizace:** Tento proces je často také nazýván jako segmentace textu nebo lexikální analýza. Při tokenizaci převádíme skupinu vět na tokeny. Nejjednodušší je si to uvést na příkladu. S věty: "Mars je podobná planeta Zemi. Můžeme zde přežít?" dostaneme po tokenizaci: 'Mars' 'je' 'podobná' 'planeta' 'Zemi' '.' 'Můžeme' 'zde' 'přežít' '?'.

- **Normalizace:** Jak je již patrné z posledního kroku tokenizace, tak výsledek tohoto kroku nemůžeme považovat za finální. Potřebujeme získat ještě smysluplnější data, která nebudou obsahovat slova v nichž jsou uložená jen interpunkční znaménka, zastavovací slova (je, v, může, atd...), slova s velkými a malými písmeny. Využíváme procesů **stemmingu** a **lemmatizace**. Po nich dostáváme z předchozích vět: 'mars' 'podobná' 'planeta' 'zemi' 'zde' 'přežít'. 

Nejprve si naimportujeme knihovny, které jsou nutné pro různé kroky nebo zrychlují práci s daty
 

```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfTransformer
from collections import Counter
import scipy.sparse as sp
from numpy.linalg import norm
from sklearn.feature_extraction.text import CountVectorizer
from collections import defaultdict 
import spacy
```

V tomto příkladu si ukážeme práci s metodou Bag of Words. Pro jednoduchost příkladu si ukážeme práci s 8 dokumenty. Každý z nich je složen pouze z jediné věty (v anglickém jazyce). Každý z dokumentů je ještě označen štítkem odpovídající, jakému tématu se věnuje. 

```python
corpus = [
    'The sky is blue and beautiful.',
    'Love this blue and beautiful sky!',
    'The quick brown fox jumps over the lazy dog.',
    "A king's breakfast has sausages, ham, bacon, eggs, toast and beans",
    'I love green eggs, ham, sausages and bacon!',
    'The brown fox is quick and the blue dog is lazy!',
    'The sky is very blue and the sky is very beautiful today',
    'The dog is lazy but the brown fox is quick!'
]
labels = [
    'weather', 'weather', 'animals', 'food', 'food', 'animals', 'weather',
    'animals'
]

corpus = np.array(corpus)
corpus_df = pd.DataFrame({'Dokument': corpus, 'Kategorie': labels})
corpus_df
```

Jak již víme, než začneme pracovat s daty, musíme provést předzpracování textu. Obvykle bychom provedly všechny kroky, abychom měli jistotu, že data jsou v takovém stavu, který nám vyhovuje pro další zpracování. My můžeme v tomto případě některé kroky vynechat, díky tomu, že víme jak datový soubor vypadá. Pro jednoduchost provedeme všechny kroky v jedné funkci.

```python
nltk.download('stopwords')  #stahování stopwords
wpt = nltk.WordPunctTokenizer() #tokenizace slov
stop_words = nltk.corpus.stopwords.words('english') #určení stopwords pro anglická slova


def normalize_document(doc):
    # nahrazení znaků na malá písmena
    # odstranení speciálních znaků/whitespaců
    doc = re.sub(r'[^a-zA-Z\s]', '', doc, re.I | re.A) #re.A (ASCII-only matching), re.I (ignore case), | OR
    doc = doc.lower() #malá písmena
    doc = doc.strip() #odstranění přebytečných mezer atd..

    # tokenizace dokumentů
    tokens = wpt.tokenize(doc)

    # odstranění stopwords ze všech dokumentů
    filtered_tokens = [token for token in tokens if token not in stop_words]

    # znovuvytvoření dokumentu z vyfiltrovaných tokenů
    doc = ' '.join(filtered_tokens)
    return doc


normalize_corpus = np.vectorize(normalize_document)
```

```python
#tento výpis slouží jen pro výpis rozdílu mezi normálním a normalizovaným textem
norm_corpus = normalize_corpus(corpus)
print(corpus)
print("="*50)
print(norm_corpus)
```

**Model Bag of Words**

Model Bag of Words je jedním z nejjednoduších pro reprezentaci textu v číselné podobě vektorů. Při získávání informací z tohoto modelu předpokládáme, že u dokumentu ignorujeme jeho slovosled, gramatiku, syntaxi a další faktory. Považujeme ho tak za soubor téměř náhodných a neuspořádaných slov. Umístění každého slova v dokumentu je nezávislé na tom, kde se objevují jiná slova. Věty jsou transformovány na vektory za pomoci funkce CountVectorizer(). Ta transformuje text na řídkou matici. Jedinečná slova jsou reprezentovány sloupcemi a řádek reprezentuje počet slov. 

Metodu BoW lze považovat za statistický histogram. Používá se především pro vyhledávání textu, klasifikaci dokumentů a zpracování. 

```python
cv = CountVectorizer(min_df=0., max_df=1.)
cv_matrix = cv.fit_transform(norm_corpus)
cv_matrix
```

```python
#reprezentace hodnot
print(cv_matrix)
```

```python
#reprezentace v podobě matice
cv_matrix = cv_matrix.toarray()
cv_matrix
```

**TF-IDF**

Další z často využívaných metod je TF-IDF. Tato technika narozdíl od jednoduché BoW, která má již známé problémy a hlavně ten, že zachází se všemi slovy stejně. Což znamená, že nedokáže rozlišit velmi běžná slova od těch vzácných (málo se vyskytujícíh). Právě ten dokáže řešit TF-IDF. Ta přiřazuje jednotlivým slovům tzv. míru důležitosti podle toho, jak často se vyskytuje v dokumentu. 

Nejprve si vysvětlíme oba pojmy, ze kterých se název metodiky skládá.

 - TF (Term Frequency) = označuje frekvenci slova v dokumentu. Pro zadané slovo definuje poměr počtu výskytů vůči celkovému počtu slov v dokumentu. 
 - IDF (Inverse document frequency) = měří důležitost slova ve všech dokumentech v korpusu (korpus = sbírka všech dokumentů v datové sadě). Pomáhá to určit vzácná slova, která se nachází ve více dokumentech.
 


```python
tt = TfidfTransformer(norm='l2', use_idf=True, smooth_idf=True)
tt_matrix = tt.fit_transform(cv_matrix)

tt_matrix = tt_matrix.toarray()
vocab = cv.get_feature_names()
pd.DataFrame(np.round(tt_matrix, 2), columns=vocab)
```

```python
# získání unikátních slov z dat
unique_words = list(
    set([word for doc in [doc.split() for doc in norm_corpus]
         for word in doc]))

# vytvoření slovníku 
def_feature_dict = {w: 0 for w in unique_words}

print('Unikátní slova:', unique_words)
print('Slovník:', def_feature_dict)
```

```python
# pro každý dokument z korpusu to vytvoří počty obsažených slov
bow_features = []
for doc in norm_corpus:
    bow_feature_doc = Counter(doc.split())
    # Vytvoření slovníku pro korpus
    all_features = Counter(def_feature_dict) 
    
    # Aktualizace slovníku pro každý dokument (řádek)
    bow_feature_doc.update(all_features)
    
    # připojení aktualizovaného slovníku
    bow_features.append(bow_feature_doc)

bow_features = pd.DataFrame(bow_features)
bow_features
```

```python
feature_names = list(bow_features.columns)

# Vytvoří tabulku vykazující počty slov pro celý korpus
df = np.diff(sp.csc_matrix(bow_features, copy=True).indptr)

df = 1 + df  # adding 1 to smoothen idf later

# show smoothened document frequencies
pd.DataFrame([df], columns=feature_names)
```

```python
# Výpočet IDF hodnoty pro každý termín/slovo
total_docs = 1 + len(norm_corpus)
idf = 1.0 + np.log(float(total_docs) / df)

# show smoothened idfs
pd.DataFrame([np.round(idf, 2)], columns=feature_names)
```

```python
# výpočet TF/IDF matice pro každý dokument
tf = np.array(bow_features, dtype='float64')
tfidf = tf * idf

pd.DataFrame(np.round(tfidf, 2), columns=feature_names)
```

```python
# Výpočet L2 normy, tento postup se využívá pokud chceme 
# dostat normalizovaný výpočet TF/IDF
norms = norm(tfidf, axis=1)
# Získání normalizace pro každý dokument
print(np.round(norms, 3))
```

```python
# Výpočet normalizace
norm_tfidf = tfidf / norms[:, None]

# Výsledna TF/IDF normalizovaná matice pro každý dokument
pd.DataFrame(np.round(norm_tfidf, 2), columns=feature_names)
```

**Word2Vec**

Word2Vec byl představen v roce 2013 týmem výzkumníků z Google. Implementace byla v C. Do Pythonu převedeno Gensimem.

Základním předpokladem Word2Vec je, že dvě slova sdílející podobné kontexty také sdílejí podobný význam a následně podobnou vektorovou reprezentaci z modelu. Například: "pes", "štěně" a "štěňátko" se často používají v podobných situacích, s podobnými okolními slovy jako "hodný", "huňatý" nebo "roztomilý", a podle Word2Vec proto budou sdílet podobnou vektorovou reprezentaci.

Z tohoto předpokladu lze Word2Vec použít ke zjištění vztahů mezi slovy v datové sadě, k výpočtu podobnosti mezi nimi nebo k použití vektorové reprezentace těchto slov jako vstupu pro jiné aplikace, jako je klasifikace textu nebo shlukování.

```python
df = pd.read_csv('https://akela.mendelu.cz/~xhubeny/UI/simpsons_dataset.csv')
df.shape
```

```python
df.head()
```

```python
#kolik hodnot je nul a jejich odstranění
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df.isnull().sum()
```

```python
#spaCy je open-source softwarová knihovna pro pokročilé zpracování přirozeného jazyka.
#nlp = spacy.load('en_core_web_sm', disable=['ner', 'parser']) # pro zrychlení
nlp = spacy.load('en_core_web_sm') 

def cleaning(doc):
    #Lematizace a odstranění Stopwords
    txt = [token.lemma_ for token in doc if not token.is_stop]

    #Bere v uvahu pouze vety delší než 2 slova
    if len(txt) > 2:
        return ' '.join(txt)
```

```python
brief_cleaning = (re.sub("[^A-Za-z']+", ' ', str(row)).lower() for row in df['spoken_words'])
```

```python
#***DLOUHÉ***

#Využití atributu spaCy .pipe() k urychlení procesu čištění:
txt = [cleaning(doc) for doc in nlp.pipe(brief_cleaning, batch_size=5000)]
```

```python
#Umístěte výsledky do DataFrame, abyste odstranili chybějící hodnoty a duplikáty:
df_clean = pd.DataFrame({'clean': txt})
df_clean = df_clean.dropna().drop_duplicates()
df_clean.shape
```

```python
#Balíček Gensim Phrases používáme k automatické detekci běžných frází (bigramů) ze seznamu vět.
#Hlavním důvodem, proč to děláme, je chytit slova jako "mr_burns" nebo "bart_simpson"!
from gensim.models.phrases import Phrases, Phraser
```

```python
#jako vstup vezme seznam slov
sent = [row.split() for row in df_clean['clean']]
```

```python
#výtváří fráze, musí se objevit alespoň 30krát, a bere to vstupy po 10 000
phrases = Phrases(sent, min_count=30, progress_per=10000)
```

```python
#Phraser sníží spotřebu paměti
bigram = Phraser(phrases)
```

```python
#transformace korpusu na základě zjištěných frází
sentences = bigram[sent]
```

```python
#Výpis nejčastěji se vyskytujících slov
#provádíme to za účelem kontroly jak dobře se nám povedlo předzpracování textu
#lemmatizace, stopwords, bigramy

word_freq = defaultdict(int)
for sent in sentences:
    for i in sent:
        word_freq[i] += 1
len(word_freq)
```

```python
sorted(word_freq, key=word_freq.get, reverse=True)[:10]
```

```python
#Trénování modelu, implementace Word2Vec od Gensimu
import multiprocessing

from gensim.models import Word2Vec
```

```python
cores = multiprocessing.cpu_count() # Count the number of cores in a computer
```

```python
# V prvním kroku nastavujeme parametry modelu jeden po druhém
# Ve druhém sestavíme slovník ze sledu vět a tím inicializuje model
# Nakonec trénujeme model
w2v_model = Word2Vec(min_count=20, #ignoruje slova s absolutní frekvencí nižší než číslo
                     alpha=0.03, #počáteční rychlost učení
                     min_alpha=0.0007, #Rychlost učení bude lineárně klesat na min_alpha s postupem tréninku
                     workers=cores-1) #počet vláken pro trénování
```

```python
#t = time()

w2v_model.build_vocab(sentences, progress_per=10000)

#print('Time to build vocab: {} mins'.format(round((time() - t) / 60, 2)))
```

```python
#*** DLOUHE ***

w2v_model.train(sentences, total_examples=w2v_model.corpus_count, epochs=30, report_delay=1)
```

## Praktický příklad generování obrazu na základě textu za využití Stable Diffusion

Pro náš příklad využijeme modelu *Stable diffusion V1* (https://huggingface.co/CompVis/stable-diffusion-v-1-4-original), jedná se o model vytvořený společnostmi CompVis, Stability AI a LAION. Pro jeho trénink byly využity obrázky z databáze LAION-5B ("*Five Billion Image-Text Pair Dataset*") a rozměrech 512x512. Jakožto minimální požadavek pro spuštění modelu se uvádí GPU s minimálně 10GB VRAM.

**Slabá místa modelu:** modely v obecné rovině mohou podporovat společenské předsudky, jelikož jsou přímo závislé na na datech, na kterých byly trénovány. Jelikož byl náš model trénován na subsetu LAION-2B(en), který obsahuje obrázky s anglickými popisky, tak při generování obrázků nebudou rovnoměrně zastoupeny obrázky a texty kultur nehovořících anglicky.
Toto vede k tomu, že ve výstupech se výrazně častěji objevují obrázky ze západních kultur. Taktéž je model výrazně horší, pokud by měl generovat na základě neanglického textu (s češtinou si neporadí ani v jednoduchých případech jako "muž").
