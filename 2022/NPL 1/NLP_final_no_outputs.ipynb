## Predspracovanie textu
Predspracovanie textových dát je postupnosť určitých krokov, ktorými pripravujeme analyzovaný text pre požadovanú úlohu. Predspracovanie hrá vo väčšine prípadov
akejkoľvek úlohy z oblasti NLP veľmi dôležitú úlohu. Jeho hlavným cieľom je zvýšiť presnosť modelu, urýchliť proces trénovania/používania a znížiť výpočtovú náročnosť
danej úlohy. Predspracovanie textu zahrňuje napríklad rozdelenie textu na menšie
jednotky, odstránenie stop slov, odstránenie interpunkcie, konverzia všetkých veľkých písmen na malé a ďalšie pokročilejšie metódy. Taktiež je nutné podotknúť,
že rôzne úlohy z oblasti NLP si vyžadujú iné formy predspracovania, ba v určitých
prípadoch to môže mať na výsledok až negatívny vplyv.
Ako bolo už spomínané, príprava textu sa skladá z viacerých krokov a to napríklad:

*   Tokenizácia
    *    Pred použitím je potrebné celý textový korpus rozložiť na individuálne elementy,
ktoré následne slúžia ako vstup pre rôzne NLP algoritmy. V mnohých prípadoch je
tokenizácia spojená s inými voliteľnými krokmi ako odstránenie interpunkcie, odstránenie špeciálnych znakov, case-folding apod.
*   Stopwords
    *    V texte sa nachádza množstvo slov, ktoré nedržia žiaden informatívny význam
a v texte sa vyskytujú najfrekventovanejšie. Tieto slová sa nazývajú stop slová.
Stop slová sú funkčné slová, závislé na danej jazykovej doméne, ktoré nenesú žiadnu
významnú informáciu. Jedná sa hlavne o slovné druhy ako zámená, predložky, spojky a podobne. Zostavenie zoznamu týchto slov sa líši v závislosti od riešenej úlohy.
*    Stematizácia
    *    Stematizácia je proces, pri ktorom je slovo transformované
na jeho slovný koreň. Nakoľko sa pri
tomto procese odstraňuje morfologická koncovka, prípadne predpona slova, po
aplikácií stematizácie môžeme dostať slová, ktoré nie sú reálne.
*    Lematizácia
    *    Lematizácia je podobný proces ako stematizácia, no na rozdiel
od stematizácie je jeho cieľom získavať nie kmeň slova, ale jeho gramaticky
správnu formu ktorá sa nazývanú lemma. Cieľom lematizácie a taktiež stematizácie je zredukovať počet unikátnych slov v slovnej zásobe, čo môže viesť
k väčšej efektivite trénovaného modelu. 
*    Part-Of-Speech
    *    POS tagging je proces, pri ktorom je ku každému slovu vo vete priradený jeho vetný člen (podstatné meno,
sloveso, zámeno a pod.) na základe jeho definície a kontextu. POS Tagging
sa v NLP okrem analýzy sentimentu taktiež používa pri rozpoznávaní pomenovaných entít (NER) alebo odpovedaní na otázky (Question answering)


## Feature extraction
Drvivá väčšina algoritmov strojového učenia nedokáže spracovať vstup vo forme
obyčajného textu. Preto je potrebné, aby sme text dokázali reprezentovať formou,
ktorá je počítaču zrozumiteľnejšia a použiť takto reprezentované dáta ako vstup
pre náš algoritmus. Aby bol daný algoritmus schopný spracovať požadovaný text,
je nutné text prekonvertovať do numerickej podoby, a to konkrétne do číselných
vektorov. Tento proces sa v NLP nazýva feature extraction.

### Bag-Of-Words
Medzi jednu z najpopulárnejších a najjednoduchších metód reprezentácie dát patrí
takzvaný bag-of-words model (BoW). Prvým krokom BoW modelu je zostavenie slovnej zásoby, ktorá je zložená zo všetkých unikátnych slov v danom korpuse. Ako už názov napovedá („Bag”), jedná
sa o pomyselné vrece, nakoľko všetky informácie o poradí alebo štruktúre slov v danom dokumente sú stratené. Jediné, čo nás pri tomto prístupe zaujíma, je to, či sa slovo z danej slovnej zásoby v dokumente nachádza alebo nie, prípadne koľkokrát
sa v dokumente nachádza.
Základnou myšlienkou modelu je predpoklad, že dokumenty, ktoré obsahujú rovnaké slová, sú podobné. Použitie BoW modelu môže fungovať nasledovne.

Ako príklad uvedieme dve vety, pričom nám každá veta reprezentuje jeden
dokument:
*    D1: **„S fotoaparátem jsem spokojenej.”**
*    D2: **„S výdrží baterie jsem spokojenej.”**

V tomto prípade bude slovná zásoba pozostávať zo šiestich slov. 

![bag-of-words.png](https://akela.mendelu.cz/~xmikula1/UI/img/bag-of-words.png)

V uvedenom príklade boli na reprezentáciu každého dokumentu použité vektory s fixnou dĺžkou šiestich hodnôt. V praxi však môže slovná zásoba obsahovať tisíce
slov. To znamená, že vektor bude obsahovať veľké množstvo nulových hodnôt a budú tak tvorené riedke vektory.

### Tf-idf
Ďalším modelom pre reprezentáciu dát, ktorý rieši niektoré z nedostatkov modelu
BoW je Tf-idf model. Ten na rozdiel od modelu BoW hodnotí aj ako veľmi relevantné
je slovo v dokumente s ohľadom na celý korpus.

### Word2vec
Word2vec model je založený na neurónovej sieti. Pri starších textminingových prístupoch nastáva problém s vysokou dimenzionalitou dát. Pri veľkom počte unikátnych slov, kde každý stĺpec reprezentuje jedno dané slovo a každá hodnota v stĺpci reprezentuje, či sa slovo v danom texte nachádza alebo nie, môžeme
obdržať veľmi veľký počet stĺpcov. Prístup založený na Word2vec sa snaží tento
problém eliminovať a reprezentuje dáta lepším spôsobom. Word2vec používa pre
každé slovo reprezentatívny vektor, a to tak, že slová, ktoré majú podobný význam
majú aj podobnejší vektor, a naopak slová, ktoré nemajú podobné významy, majú
aj odlišné vektory. 

Tento prístup bol predstavený v dvoch metódach:
1.    **Continuous Bag-of-Words Model (CBOW)** – v tejto architektúre je snaha
o predvídanie daného slova na základe slov, ktoré sú v texte pred a za slovom,
ktoré sa snažíme určiť (jeho kontext). Na základe viacerých vstupov získavame
jeden výstup (Ayyadevara, 2018).
2.    **Continuous Skip-gram Model** – na rozdiel od CBOW architektúry
Skip-gram model nepredvída dané slovo na základe jeho kontextu, ale snaží
sa určiť susedné slová, ktoré by sa mohli v texte nachádzať na základe zadaného slova. Na základe jedného vstupu získavame viacero výstupov

### BERT model
**Transformer model**

Základným stavebným kameňom modelu BERT je Transformer model, ktorý
sa dokáže naučiť kontextové väzby medzi slovami v texte. Transformer model obsahuje dva separátne mechanizmy. **Encoder**, ktorý spracuje textový vstup a **decoder**,
ktorý produkuje výstup pre požadovanú úlohu. Pretože hlavnou úlohou modelu
BERT je vytvoriť určité porozumenie jazyka, vystačí si s použitím iba encoder mechanizmu. Vstupom pre encoder je sekvencia tokenov, ktoré sú najprv vložené do
vektorov a následne spracované v neurónovej sieti. Výstupom je množina vektorov,
v ktorej každý vektor reprezentuje jeden vstupný token.

![transformers.png](https://akela.mendelu.cz/~xmikula1/UI/img/transformers.png)

Hlavnou inováciou modelu BERT je použitie bidirectional trénovania, čo znamená, že na rozdiel od predchádzajúcich modelov, ktoré spracovávali vstupnú textovú sekvenciu buď zľava doprava alebo kombináciou spracovania textu zľava doprava
a sprava doľava, dokáže spracovávať celú vstupnú sekvenciu slov naraz. To umožňuje
modelu lepšie pochopiť kontexty jednotlivých slov s ohľadom na okolie. Pri trénovaní mnohých jazykových modelov je cieľom predpovedať nasledujúce slovo v danej
vete. Napríklad vo vete: **„Mobil má vynikajúci ___,“** sa model snaží predvídať,
čo bude nasledovať za slovom vynikajúci. Takáto forma trénovania značne limituje
model pri porozumení kontextu.
BERT preto predstavuje dve formy unsupervised tréningu, ktoré riešia tento
problém:
*    **Masked Language Model** – pred predložením vstupných sekvencií modelu
sa náhodne zamení 15 % tokenov každej vstupnej sekvencii tokenom \[MASK\],
pričom cieľom je zistiť originálne slová iba na základe kontextu, ktorý poskytujú
ostatné nemaskované slová v sekvencii. To umožňuje modelu zlúčiť kontext na
pravej aj ľavej strane.
*    **Next Sentence Prediction** – preto, aby bol model schopný dokázať rozpoznať
vzťahy medzi jednotlivými vetami, bol model trénovaný za účelom binárneho
rozpoznávania toho, či z dvoch viet, A a B, ktoré máme na vstupe, nasleduje
veta B za vetou A alebo nie. Pri výbere týchto viet bola v polovici zo všetkých
príkladov veta B reálnou nasledujúcou vetou za vetou A a v druhej polovici
bola z korpusu vybraná náhoda veta.

*   Fine-tuning
BERT môže byť použitý na širokú škálu špecifických jazykových úloh pri pridaní
len jednej výstupnej vrstvy. Pri klasifikačných úlohách, ako analýza sentimentu,
stačí pridať na výstup encoderu jednu klasifikačnú vrstvu.



### GPT-3
#### GPT v skratke
Generative Pre-trained Transformer (GPT) modely od spoločnosti OpenAI sú v poslednej dobe horúcou novinkou v NLP komunite, nakoľko priniesli veľmi sofistikované a výkonné jazykové modely. Tieto jazykové modely môžu byť využité napr. na question answering, textual entailment, text summarisation, atď. a to bez učenia sa s učiteľom. Tieto jazykové modely potrebujú veľmi málo, alebo žiadne, príklady, aby pochopili zadanie úlohy a vykonali analýzu porovnateľnú, ak nie lepšiu, ako tie najlepšie modely trénované s učiteľom.

GTP modely prešli v posledných 2 rokoch viacerými zmenami a vylepšeniami, ktoré vytvorili nasledujúce verzie:
1. **GPT-1** (Jazykové porozumenie prostredníctvom generatívneho predtrénovania)
2. **GPT-2** (Jazykové modely sú multitaskingový študenti bez dohľadu)
3. **GPT-3** (Jazykové modely sú "few shots" študenti)

#### GPT-3
Open AI sa podarilo vytvorením modelu GPT-3 vybudovať na základoch architektúry transformátorov komplexný a silný prediktívny autoregresívny jazykový model, ktorý používa hĺbkove učenie. Dokáže vykonávať úlohy bez potreby ladenia parametrov a len na základe malého/žiadneho trénovacieho setu (pre predstavu GPT-3 model obsahuje viac ako 175 miliárd parametrov, čo je 10x viac ako známy Turingov NLG jazykový model od Microsoftu). Vďaka veľkému množstvu parametrov a rozsiahlemu trénovaciemu setu, na ktorom bol GPT-3 natrénovaný, funguje dobre pri následných úlohách NLP v nastavení zero-shot a few-shot. Vďaka svojej veľkej kapacite má schopnosti ako napr. písanie článkov, ktoré je ťažké odlíšiť od článkov napísaných ľuďmi. Môže tiež vykonávať úlohy, na ktoré nebol nikdy výslovne trénovaný, ako je napr. sčítanie čísel, písanie SQL dotazov a kódov, dešifrovanie slov vo vete, písanie kódov v Reacte alebo JavaScript s popisom úlohy v prirodzenom jazyku a podobne.

##### **Učebné ciele a koncepty**

<u>Učenie sa v kontexte</u> 

Veľké jazykové modely si na zákalde textových dát z trénovacieho setu vytvoria nástroje a ďalšie zručnosti pre rozpoznávanie vzorov. Zatiaľ čo sa jazykové modely učia primárnemu cieľu predpovedať ďalšie slovo s danými kontextovými slovami, začínajú rozpoznávať vzory v údajoch, ktoré im pomáhajú minimalizovať straty. Neskôr táto schopnosť (rozpoznávanie vzorov) pomáha modelu pri prenose úloh typu zero-shot. Keď sú jazykové modely prezentované s niekoľkými príkladmi a/alebo popisom toho, čo musí model urobiť, zhodujú sa so vzorom príkladov s tým, čo sa naučil v minulosti pre podobné údaje, a tieto znalosti využívajú na vykonávanie daných úloh. Ide o silnú schopnosť veľkých jazykových modelov, ktorá sa zvyšuje so zvyšujúcim sa počtom parametrov modelu.

<u>Nastavenie few-shot, one-show alebo zero-shot</u> 

Few, one a zero show nastavenia sú špeciálnymi prípadmi zero-shot transféru. Vo few-shot nastavení je modelu poskytnutý popis úlohy s toľkým množstvom príkladov, ako je možné vložiť do kontextového poľa modelu. Vo one-shot nastavení je modelu poskytnutý len jeden príklad a v zero-shot nastavení ani jeden.

##### **Ako model GPT-3 funguje**
Ako už bolo spomenuté ide o prediktívny jazykový model, čo znamená, že vuyžíva model strojového učenia neurónovej siete, ktorý na vstupe príjme text a dokáže ho transformovať na to, čo predpovedá, že bude najužitočnejší výsledok. ##### **Výhody GPT-3 modelu**
GPT-3 poskytuje dobré riešenie vždy, keď je potrebné vygenerovať veľké množstvo výstupného textu na základe malého množstva vstupného textu. Existuje mnoho situácií, v ktorých nie je praktické alebo efektívne mať po ruke človeka na generovanie textového výstupu, alebo môže byť potrebné automatické generovanie textu, ktorý sa javí ako ľudský. Napríklad centrá služieb zákazníkom môžu používať GPT-3 na zodpovedanie otázok zákazníkov alebo na podporu chatbotov; obchodné tímy ho môžu použiť na spojenie s potenciálnymi zákazníkmi; a marketingové tímy môžu písať detaily analýz pomocou modelu GPT-3.

##### **Nevýhody/limity GPT-3 modelu**
Najväčším problémom je, že sa GPT-3 model neustále neučí. Bol vopred natrénovaný, čo znamená, že nemá nepretržitú dlhodobú pamäť, ktorá sa učí z každej interakcie. Okrem toho model GPT-3 trpí rovnakými problémami ako všetky neurónové siete: ich nedostatok schopnosti vysvetliť a interpretovať, prečo určité vstupy vedú k špecifickým výstupom.

Okrem toho architektúry transformátorov - medzi ktoré patrí aj GPT-3 - trpia problémami s obmedzenou veľkosťou vstupu. Používateľ nemôže poskytnúť veľa textu ako vstup pre výstup, čo môže obmedziť niektoré aplikácie. GPT-3 si konkrétne dokáže poradiť len so vstupným textom v dĺžke niekoľkých viet. GPT-3 tiež trpí pomalým časom vyvodzovania, pretože trvá dlho, kým model vygeneruje výsledok.

Ukážka podobností viet na základe získaných embeddingov. Vektory vstupného textu sú vytvárané tak, že vektory textov s vysokovou podobnosťou sú vo vektorom priestore bližšie a texty s malou podobnosťou sú naopak vzdialené. 
To nám umožňuje jednoducho zistiť podobnosť daných textov napríklad s použitím kosínovej podobnosti. 

```python
# imports
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# load model -> all-MiniLM-L6-v2
# This is a sentence-transformers model: It maps sentences & paragraphs 
# to a 384 dimensional dense vector space and can be used for tasks 
# like clustering or semantic search.
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# example sentences
sentences = ["I ate dinner.", 
       "We had a three-course meal.", 
       "Brad came to dinner with us.",
       "He loves fish tacos.",
       "In the end, we all felt like we ate too much.",
       "We all agreed; it was a magnificent evening."]

# get embeddings for example sentences
sentence_embeddings = model.encode(sentences)

# query on which we gonna compare similarities
query = "I had pizza and pasta"

# encode query sentence
query_vec = model.encode([query])[0]

# get similarities between query sentence and example
# sentences and store them into dataframe
my_dict = {}
sims = []

for sent in sentences:
    sims.append(round(util.cos_sim(query_vec, model.encode([sent])[0])[0].item(), 4))

my_dict['Sentece'] = sentences
my_dict['Similarity'] = sims 

df_sim = pd.DataFrame(my_dict)
print('Query sentence: ', query)
display(df_sim)
```

```python
sentence = 'This framework generates embeddings for each input sentence'

# get embedding for example sentence
embeddings = model.encode([sentence])

#Print embedding
print("Sentence:", sentence)
print("Embedding shape:", embeddings[0].shape)
print("Embedding:", embeddings[0])
```

```python
# set max lenght of tokens, in this example max len of the model
MAX_LEN = 512
```

```python
# get encodings for train set
train_encodings = tokenizer(train_texts, truncation=True, padding='max_length', max_length=MAX_LEN)
```

```python
# get encodings for eval set
val_encodings = tokenizer(val_texts, truncation=True, padding='max_length', max_length=MAX_LEN)
```

```python
# get encodings for test set
test_encodings = tokenizer(test_texts, truncation=True, padding='max_length', max_length=MAX_LEN)
```

```python
# create MyDataset class that inherit from PyTorch Dataset class
class MyDataset(torch.utils.data.Dataset):
    # initilize data and their labels
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    # returns dictionary with input ids (tokens ids),
    # attention mask and label tensor
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
    
    # should return len of dataset
    def __len__(self):
        return len(self.labels)

# create datasets from encodings and their labels
train_dataset = MyDataset(train_encodings, train_labels)
val_dataset = MyDataset(val_encodings, val_labels)
test_dataset = MyDataset(test_encodings, test_labels)
```

```python
# example of one instance represented in our custom MyDataset
train_dataset[0]
```

```python
# define function for compuning custom metrics by which to evaluate our model
def compute_metrics(pred):
    # real labels
    labels = pred.label_ids

    # predictied labels
    preds = pred.predictions.argmax(-1)
    
    # compute metrics
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

```python
# load pre-trained model on which we are gonna perform fine-tuning
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")
```

```python
# initilize training parameters for Trainer class
training_args = TrainingArguments(
    output_dir='./results',
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=1000,
    save_strategy='no',
    logging_strategy='epoch',
)
```

```python
# move model to GPU for better performance
model.to(device)
```

```python
# initilize Trainer class
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)
```

```python
# train model
trainer.train()
```

```python
# evaluate model on evaluation set
trainer.evaluate()
```

```python
# get predictions on test test
predictions = trainer.predict(test_dataset=test_dataset)
```

```python
# custom fucntion that return sentiment predicted by our model on input text
def get_sentiment(text):
    input = tokenizer.encode(text, truncation=True, padding=True, return_tensors="tf")
    output = model.predict(input)[0]
    prediction = tf.nn.softmax(output, axis=1).numpy()[0]
    
    if np.argmax(prediction)==1:
        print("Recenzia: ", text)
        print("Sentiment: POSITIVE")
    else:
        print("Recenzia: ", text)
        print("Sentiment: NEGATIVE")
```
